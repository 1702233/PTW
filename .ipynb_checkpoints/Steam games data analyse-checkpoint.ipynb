Dat was nog niet zo moeilijk. Laten we nu proberen dit te sortern op prijs net zoals eerder.

```python
def sort_func(x):
    return x['price']

data.sort(key=sort_func, reverse=True)

data[:3]
```

Dit is opzich prima te doen, maar koste mij veel meer tijd om voor elkaar te krijgen door meer zoek werk. Laten we nu een eigen sorteer algoritme gebruiken.


```python
def swap(lst, i, j):
    x = lst[i]
    y = lst[j]

    lst[i] = y
    lst[j] = x

def my_sort(lst, key=None):  
    if key:
        pass
    else:

        steps = 0

        while True:
            swapped = False

            for i in range(len(lst)-1):
                steps += 1
                if lst[i] > lst[i+1]:
                    swap(lst, i, (i+1))
                    swapped = True

            if not swapped:
                break

    return steps
```

```python
my_sort([5, 8, 3, 13, 1, 6])
```
